### Package Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Algorithm Realization

```python
class LinearRegression(object):
    """
    A class object for using a linear regression model.
    Besides general realization, i add l1-normalization and l2-normalization, 
    which can be selected in initialization.
    """
    def __init__(self, penalty='null', learning_rate=0.01, iteration=200, lamda=0.1):
        self.penalty = penalty
        self.learning_rate = learning_rate
        self.iteration = iteration
        self.theta = None
        self.lamda = lamda
        
    def fit(self, X, y):
        self.theta, Js = self.gradient_descent(X, y)
        return Js
        
    def gradient_descent(self, X, y):
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        theta = np.zeros((X.shape[1],1))
        m = X.shape[0]
        Js = [self.compute_cost(X, y, theta)]
        
        for i in range(self.iteration):
            penalty = 0
            if self.penalty == 'l2':
                theta_bac = np.concatenate(([[0],], theta[1:]),axis=0)
                penalty = self.lamda * theta_bac
            elif self.penalty == 'l1':
                theta_bac = np.sign(theta)
                theta_bac[0] = 0
                penalty = self.lamda * theta_bac
                
            grad = X.T@(X@theta - y) + penalty
            theta -= self.learning_rate * grad/m
            Js.append(self.compute_cost(X, y, theta))
        return theta, Js
    
    def predict(self, X):
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        y = X@self.theta
        return y
            
    def compute_cost(self, X, y, theta):
        m = X.shape[0]
        J = 1/(2*m) * (X@theta - y).T @ (X@theta - y)
        penalty = 0
        if self.penalty == 'l2':
            penalty = 1/(2*m) * self.lamda * np.sum(theta[1:]**2)
        elif self.penalty == 'l1':
            penalty = 1/m * self.lamda * np.sum(np.abs(theta[1:]))
        
        return J+penalty
```

### Test example

```python
# input data x, label data y
X = np.array([[1],
     [2],
     [3],
     [4]])
y = np.array([[1],
     [2],
     [3],
     [4]])
```

```python
# test
lr = LinearRegression()
costs1 = lr.fit(X,y);
theta1 = lr.theta
costs1 = [float(cost) for cost in costs1]

lr = LinearRegression(penalty='l1', lamda=10)
costs2 = lr.fit(X,y);
theta2 = lr.theta
costs2 = [float(cost) for cost in costs2]

lr = LinearRegression(penalty='l2', lamda=10)
costs3 = lr.fit(X,y);
theta3 = lr.theta
costs3 = [float(cost) for cost in costs3]
```

```python
x = np.arange(201)
plt.plot(x, costs1, '--', color='b', label='cost function')
plt.plot(x, costs2, '--', color='r', label='cost function_l1(lambda=10)')
plt.plot(x, costs3, '--', color='g', label='cost function_l2(lambda=10)')
plt.legend();
```

```python
x = np.linspace(0, 4, 100)
y1 = x * theta1[1] + theta1[0]
y2 = x * theta2[1] + theta2[0]
y3 = x * theta3[1] + theta3[0]

plt.plot(x, y1, '--', color='b', label='pred line')
plt.plot(x, y2, '--', color='r', label='pred line_l1(lambda=10)')
plt.plot(x, y3, '--', color='g', label='pred line_l2(lambda=10)')
plt.legend();
```
